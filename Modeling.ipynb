```python
import pandas as pd
import numpy as np
import sklearn
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/housing_data_08-02-2021.csv')

df['size'] = pd.to_numeric(df['size'], errors='coerce')
df['price'] = pd.to_numeric(df['price'], errors='coerce')

from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='median')

df_fillna = imputer.fit_transform(df[df.columns[2:]])

housing = pd.DataFrame(df_fillna,columns=df.columns[2:],index=df.index)
```

```python
from sklearn.model_selection import train_test_split

train_set , test_set = train_test_split(housing,test_size=0.2,random_state=35)
```

```python
X_train = train_set.drop('price',axis=1)
y=train_set['price'].copy()
```

```python
from sklearn.preprocessing import StandardScaler

full_pipeline = StandardScaler()
```

```python
X_prepared = full_pipeline.fit_transform(X_train)
```

MODELING
LinearRegression

```python
from sklearn.linear_model import LinearRegression

LR_model = LinearRegression()
```

```python
LR_model.fit(X_prepared,y)
```

```python
test_data = X_train.sample(10)
test_data
```

```python
test_label = y.loc[test_data.index]
test_label
```

```python
test_data_prepared = full_pipeline.transform(test_data)
test_data_prepared
```

```python
predicted_data = LR_model.predict(test_data_prepared)
predicted_data
```

```python
pd.DataFrame({'Real baho':test_label,'LR_model':predicted_data})
```

```python
X_test = test_set.drop('price',axis=1)
```

```python
y_test = test_set['price'].copy()
```

```python
X_test_prepared = full_pipeline.transform(X_test)
```

```python
y_predicted = LR_model.predict(X_test_prepared)
```

```python
from sklearn.metrics import mean_absolute_error

MAE = mean_absolute_error(y_test,y_predicted)

print('MAE=',MAE)
```

```python
from sklearn.metrics import root_mean_squared_error

RMSE = root_mean_squared_error(y_test,y_predicted)

print('RMSE=',RMSE)
```

RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor

RF_model = RandomForestRegressor()

RF_model.fit(X_prepared,y)
```

```python
y_predicted = RF_model.predict(X_test_prepared)
```

```python
MAE = mean_absolute_error(y_test,y_predicted)

print('MAE=',MAE)
```

```python
RMSE = root_mean_squared_error(y_test,y_predicted)

print('RMSE=',RMSE)
```

DECISION TREE

```python
from sklearn.tree import DecisionTreeRegressor
Tree_model = DecisionTreeRegressor()
Tree_model.fit(X_prepared, y)
```

```python
y_predicted = Tree_model.predict(X_test_prepared)
```

```python
MAE = mean_absolute_error(y_test,y_predicted)

print('MAE=',MAE)
```

```python
RMSE = root_mean_squared_error(y_test,y_predicted)

print('RMSE=',RMSE)
```
